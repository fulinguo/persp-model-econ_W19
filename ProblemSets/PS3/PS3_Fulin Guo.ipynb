### Exercise 5.1

If $T = 1$, the maximization problem to choose consumption is:

\begin{equation}\label{e1}
  \max_{c_{1}\in[0,W_{1}]} u(c_{1})  \quad s.t. \quad W_{2}=W_{1}-c_{1}
\end{equation}

Since the individual will not live in $T=2$, he/she only cares about the amount of cake he/she can consume in period 1, so the optimal choice is to consume all he/she has in $T=1$ since the utility function is increasing. Therefore, the optimal value of $c_{1}$ is $c_{1}=W_{1}$.
Equivalently, we can write the problem as:

\begin{equation}\label{e2}
  \max_{W_{2}\in[0,W_{1}]} u(W_{1}-W_{2})
\end{equation}

$W_{2}$ is the amount of cake the individual saves in period 1. Since $u(c)$ is an increasing function, the optimal value of $W_{2}$ is 0. That is, the individual will not save any cake for period 2.

### Exercise 5.2

If $T = 2$, the optimization problem is:

\begin{equation}\label{e3}
  \max_{W_{t+1}\in[0,W_{t}]} u(W_{1}-W_{2})+ \beta u(W_{2}-W_{3})
\end{equation}

Therefore, the optimal amount of cake to save for period 3 is $W_{3}=0$ since $u(c)$ is increasing in $c$ (so the lifetime utility is decreasing in $W_{3}$).
For $W_{2}$, using the fact that $W_{3}=0$ at the optimal, we could write the condition for $W_{2}$:

\begin{equation}\label{e5}
  -\frac{d u(W_{1}-W_{2})}{d c}+ \beta \frac{d u(W_{2})}{d c} = 0
\end{equation}

### Exercise 5.3

If $T = 3$, the optimization problem is:

\begin{equation}\label{e6}
  \max_{W_{t+1}\in[0,W_{t}]} u(W_{1}-W_{2})+ \beta u(W_{2}-W_{3})+ \beta^{2} u(W_{3}-W_{4})
\end{equation}

Similar to the previous questions, the condition for $W_{4}$ is $W_{4}=0$.
Then, we could write the condition for $W_{3}$:

\begin{equation}\label{e7}
  -\beta \frac{d u(W_{2}-W_{3})}{d c}+ \beta^2 \frac{d u(W_{3})}{d c} = 0
\end{equation}

The condition for $W_{2}$ is:
\begin{equation}\label{e8}
  -\frac{d u(W_{1}-W_{2})}{d c}+ \beta \frac{d u(W_{2}-W_{3})}{d c} -\beta \frac{d u(W_{2}-W_{3})}{d c} \frac{dW_{3}}{dW_{2}} +\beta^2 \frac{d u(W_{3})}{d c} \frac{dW_{3}}{dW_{2}}= 0
\end{equation}

Using the condition for $W_{3}$, we get:
\begin{equation}\label{e9}
  -\frac{d u(W_{1}-W_{2})}{d c}+ \beta \frac{d u(W_{2})}{d c} = 0
\end{equation}

If $W_{1}=1$, $\beta=0.9$, and $u(c_{t})=ln(c_{t})$, then we can get $W_{4}=0$, 
\begin{equation}\label{e10}
  -\frac {1}{W_{2}-W_{3}}+ \frac {0.9}{W_{3}} = 0
\end{equation}

Or, 
\begin{equation}\label{e11}
  W_{3}=\frac{9W_{2}}{19}
\end{equation}

And,
\begin{equation}\label{e12}
-\frac {1}{W_{1}-W_{2}} + \frac {0.9}{W_{2}-\frac {9W_{2}}{19}} = 0
\end{equation}

The solution of $W_{2}$ is:
\begin{equation}\label{e13}
W_{2}=\frac {171W_{1}}{271}=\frac{171}{271} \approx 0.631
\end{equation}

Substituting the solution of $W_{2}$ into the condition for $W_{3}$, we get $W_{3}=\frac{1539}{5149} \approx 0.2989$

Therefore, $W_{1}=1, W_{2}=0.631, W_{3}=0.2989, W_{4}=0$

Using $c_{t}=W_{t}-W_{t+1}$, we can get $c_{1}=0.3690, c_{2}=0.3321,c_{3}=0.2989$

Both $c$ and $W$ decrease with time.

### Exercise 5.4

The optimal solution for $W_{T+1}$ is 0, so the problem can be written as:

\begin{equation}\label{e14}
V_{T-1}(W_{T-1})\equiv \max_{W_{T}} u(W_{T-1}-W_{T})+\beta u(W_{T})
\end{equation}

Differentiating it w.r.t. $W_{T}$ and let $W_{T}=\psi_{T-1}(W_{T-1})$, we get:
    
\begin{equation}\label{e15}
-\frac{d u(W_{T-1}-\psi_{T-1}(W_{T-1}))}{d c}+ \beta \frac{d u(\psi_{T-1}(W_{T-1}))}{d c} = 0
\end{equation}

We can write $V_{T-1}$ in terms of $\psi_{T-1}(W_{T-1})$:

\begin{equation}\label{e16}
V_{T-1}(W_{T-1})=u(W_{T-1}-\psi_{T-1}(W_{T-1}))+\beta u(\psi_{T-1}(W_{T-1}))
\end{equation}

### Exercise 5.5

\end{equation}

### Exercise 5.6

The finite horizon Bellman equation for the function at $T-2$ is:
    
\begin{equation}\label{ee}
V_{T-2}(W_{T-2})\equiv \max_{W_{T-1}} ln(W_{T-2}-W_{T-1})+\beta V_{T}(W_{T-1})
\end{equation}

Using the envelope theorem, we get:
    
\begin{equation}\label{e19}
-\frac{d u(W_{T-2}-\psi_{T-2}(W_{T-2}))}{d c}+ \beta \frac{d u(\psi_{T-1}(W_{T-1})-\frac{\beta\psi_{T-1}(W_{T-1})}{\beta+1} )}{d c} = 0
\end{equation}

Or,

\begin{equation}\label{e20}
-\frac{1}{W_{T-2}-\psi_{T-2}(W_{T-2})}+ \frac{\beta}{(\psi_{T-1}(W_{T-1})-\frac{\beta\psi_{T-1}(W_{T-1})}{\beta+1})} = 0
\end{equation}

The analytical solution is:
\begin{equation}\label{e21}
W_{T-1}=\frac{(\beta^2+\beta)W_{T-2}}{\beta^2+\beta+1}
\end{equation}

And:
\begin{equation}\label{e22}
V_{T-2}(W_{T-2})=ln\frac{W_{T-2}}{1+\beta+\beta^2}+\beta ln\frac{\beta W_{T-2}}{1+\beta+\beta^2}+\beta^{2}ln\frac{\beta^{2} W_{T-2}}{1+\beta+\beta^2}
\end{equation}


### Exercise 5.7

The analytical solutions for $\psi_{T-s}(W_{T-s})$ and $V_{T-s}(W_{T-s})$ are:

\begin{equation}\label{e23}
\psi_{T-s}(W_{T-s})=\frac{[\sum_{i=1}^{s} \beta^{i}]W_{T-s}}{\sum_{i=0}^{s}\beta^{i}}
\end{equation}

\begin{equation}\label{e24}
V_{T-s}(W_{T-s})=\sum_{i=0}^{s} \beta^{i}ln[\frac{\beta^{i}W_{T-s}}{\sum_{i=0}^{s} \beta^{i}}]
\end{equation}

We could take the limit $s\to \infty$:

\begin{equation}\label{e25}
\lim_{s\to\infty}\psi_{T-s}(W_{T-s})=\beta W_{T-s}
\end{equation}

\begin{equation}\label{e26}
\lim_{s\to\infty}V_{T-s}(W_{T-s})=\frac{\beta ln\beta}{(1-\beta)^2}+\frac{1}{1-\beta}ln [(1-\beta)W_{T-s}]
\end{equation}

### Exercise 5.8

When the horizon is infinite, the Bellman equation is:
\begin{equation}\label{e27}
V_{T-s}(W_{T-s})=\max_{W_{T-s+1}\in[0,W_{T-s}]} u(W_{T-s}-W_{T-s+1})+\beta V_{T-s+1}(W_{T-s+1})
\end{equation}

We can omit the subscript when $T$ is infinite. That is:
\begin{equation}\label{e28}
V(W)=\max_{W^{'}\in[0,W]} u(W-W^{'})+\beta V(W^{'})
\end{equation}
Where $W^{'}$ is the amount of cake in the next period

### Exercise 5.10

```python
beta=0.9
v=np.zeros(100)
w1=np.linspace(w_min,w_max,n) # w1 is in [0.01,1], as the question mentions.
w_broad=np.tile(w.reshape(n,1),(1,n))
w1_broad=np.tile(w1,(n,1))
c_broad=w_broad-w1_broad
v_broad=np.tile(v,(n,1))
nega=(c_broad<=0)
c_broad[nega]=10**(-10)
v_broad[nega]=-10*(10)
v1=np.max(np.log(c_broad)+beta*v_broad,axis=1)
psi=[]
for i in range(n):
    psi.append(w1_broad[i][np.argmax(np.log(c_broad)+beta*v_broad,axis=1)[i]])
new_w=np.array(psi)
print('The resulting policy function is:\n', new_w)
print('The resulting value function is:\n', v1)
```

### Exercise 5.16

```python
import scipy.stats
sigma=0.5
m=7
u=4*0.5
epsilon=np.linspace(u-3*sigma,u+3*sigma,m)

gamma=np.array([0.0]*m)
for i in range(m):
    gamma[i]=scipy.stats.norm.pdf(epsilon[i],u,sigma)

print('epsilon:\n',epsilon)
print('gamma:\n',gamma)
```

### Exercise 5.17

```python
beta=0.9
w=np.linspace(w_min,w_max,n)
w1=np.linspace(w_min,w_max,n) # w1 is in [0.01, 1], as the question says
v1=np.zeros((n,m))
w_broad=np.tile(np.tile(w.reshape(n,1),(1,m)),(1,1,n)).reshape((n,m,n))
ep_broad=np.tile(np.tile(epsilon.reshape(m,1),(1,n)),(n,1,1)).reshape(n,m,n)
w1_broad=np.tile(w1,(n,m,1))
v=v1.copy()
ev=np.zeros(n)
for i in range(m):
    ev+=v.reshape(m,n)[i]*gamma[i]
ev_broad=np.tile(ev,(n,m,1))
v_broad=np.tile(v.reshape(m,n),(n,1,1))
c_broad=w_broad-w1_broad
nega=(c_broad<=0)
c_broad[nega]=10**(-10)
ev_broad[nega]=-10*(10)
v1=np.max(ep_broad*(np.log(c_broad))+beta*ev_broad,axis=2)
psi=np.zeros((n,m))
for i in range(n):
    for j in range(m):
        psi[i][j]=w1_broad[i][j][np.argmax(ep_broad*np.log(c_broad)+beta*ev_broad,axis=2)][i][j]
new_w=np.array(psi)
new_w
print('The resulting policy function is:\n',new_w)
```
